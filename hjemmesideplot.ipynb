**Box Plots over Quantiles**

```python
customPalette = sns.set_palette(sns.color_palette(USA))
sns.catplot(x="Yearly_Quantile", kind="count", palette=customPalette, data=historic_data, height=4.27, aspect=5.5/2)
plt.title("Distribution of Apartment Price Classes Sold in Manhattan from 2013-2017")
plt.xlabel("Yearly Quantile")
plt.ylabel("Number of Apartments Sold")
plt.show()
```

```python
customPalette = sns.set_palette(sns.color_palette(['pink', 'lightgreen']))
sns.catplot(x="APARTMENT TYPE", kind="count", palette=customPalette, data=historic_data, height=4.27, aspect=5.5/2)
plt.title("Distribution of Apartment Types Sold in Manhattan from 2013-2017")
plt.xlabel("Yearly Quantile")
plt.ylabel("Number of Apartments Sold")
plt.show()
```
